# Análise de Resíduos

Primeiro iremos criar variáveis a partir de uma distribuição uniforme, para a variável x.
Para a variável y, iremos criar a partir de uma distribuição de Poisson.

```r
x <- runif(200, 0.5) # 200 valores da uniforme 0.5
x
```

```r
y <- rpois(200, lambda = exp(x)) # também vamos simular 200 valores para nossa resposta e iremos escrever a média, que é o lambda dessa maneira.
y
```

Vamos fazer um gráfico para entender o comportamento de x e y

```r
plot(y ~ x, pch = 20, cex = 1.25, col = 'blue')
```

Primeiramente, vamos tentar ajustar um modelo com resposta normal

```r
glm_norm <- glm(y ~ x, family = gaussian)
par(mfrow = c(1,1))
plot(glm_norm, pch = 19, cex = 1.1)
```

No plot, temos 4 gráficos, o gráfico de resíduos vs valores ajustados, o qq-plot, os resíduos studentizados, que é a raíz quadrada dos resíduos e ainda temos o gráfico de influência, que analisa os resíduos pela distância de cook.

# Teste de normalidade dos resíduos
O teste de normalidade só será aplicado quando estivermos usando os resíduos quantílicos, pois são os únicos que vamos conseguir validar isso.

```r
shapiro.test(glm_norm$residuals)
```

```r
install.packages('statmod')
library(statmod)
```

```r
residuos <- qresid(glm_norm) #resíduos quantílicos
residuos
```

```r
ajustados <- predict(glm_norm) #valores ajustados
ajustados
```

Quando o modelo está correto, o resíduo quantílico deve acusar que tem distribuição normal.

```r
shapiro.test(residuos)
```

O resíduo não seguiu distribuição normal, logo o modelo não está bem ajustado.
Podemos realizar um gráfico de resíduos vs valores ajustados e qqplot

```r
par(mfrow = c(1,1))
plot(residuos ~ ajustados, col='blue')
qqnorm(residuos, pch = 20, cex = 1.4, col = 'blue')
qqline(residuos)
```

# Rerando um novo modelo
Estão como estamos trabalhando com contagens, vamos utilizar um GLM de poisson.


```r
glm_poisson <- glm(y ~ x, family = poisson)
glm_poisson
```

```r
par(mfrow = c(1,1))
plot(residuos ~ ajustados, col= 'blue')
qqnorm(residuos, pch = 20, cex = 1.4, col = 'blue')
qqline(residuos)
```

```r
residuos <- qresid(glm_poisson)
ajustados <- predict(glm_poisson)

shapiro.test(residuos)
```

```r
par(mfrow = c(1,1))
plot(residuos ~ ajustados, col = 'blue')
qqnorm(residuos, pch = 20, cex = 1.4, col = 'blue')
qqline(residuos)
```

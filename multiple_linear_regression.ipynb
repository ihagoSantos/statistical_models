# Gerando modelos de regressão com statsmodels

Podemos observar que o modelo conseguiu explicar 72% da variabilidade dos dados (Mostrado no R-squared). Entretanto, como é uma regressão múltipla, devemos observar o *Adj. R-squared* que fica em 71% da variabilidade dos dados.

A estatística F mostra que temos um modelo significativo.

Devemos observar o P-value para saber se as variáveis independentes tem força suficiente. As variáveis *Temperatura_Media* e *Temperatura_Minima* não foram significativas devido o problema de multicolineariedade, seria interessante remove-las.

```python
# Variáveis Independentes
X = cerveja.drop(['Consumo', 'Data'], axis=1)
# Variável Dependente
Y = cerveja['Consumo']
```

```python
modelo1 = (sm.OLS(Y, sm.add_constant(X)).fit())
modelo1.summary(title='Sumário do modelo')
```

# Análise gráfica de resíduos

```python
residuos1 = modelo1.resid
fig,ax = plt.subplots(2,2,figsize=(15,6))

residuos1.plot(title='Resíduos do modelo 1', ax = ax[0][0])
sns.distplot(residuos1, ax=ax[0][1])
plot_acf(residuos1,lags=40, ax=ax[1][0])
qqplot(residuos1,line='s', ax=ax[1][1])
```
